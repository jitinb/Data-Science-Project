# **Dataset used:**
The dataset has been taken from https://data.gov.in/.
The data set is available to be downloaded from the link https://data.gov.in/resource/all-india-consumer-price-index-ruralurban-upto-november-2021. It contains the measure of Consumer Price Indices (CPI) which changes over time in general level of prices of goods and services acquired by various households for the purpose of consumption and are widely used as a macroeconomic indicator of inflation.

````python
df=pd.read_csv("/content/drive/My Drive/Colab Notebooks/DS200/Module4/All_India_Index_july2019_20Aug2020_dec20_1_2.csv")
````

````python
df.head()
````

Taking data set for different months(January, July and December)

````python
df_Jan=df[df['Month']=='January']
df_Jul=df[df['Month']=='July']
df_Dec=df[df['Month']=='December']
````

Filtering the data sets for the respective months based on Sectors

````python
Urban_Jan=df_Jan[df_Jan['Sector']=='Urban']
Rural_Urban_Jul=df_Jul[df_Jul['Sector']=='Rural+Urban']
Rural_Dec=df_Dec[df_Dec['Sector']=='Rural']
````

# Scatter Plot

---



Making the Scatter Plot for the rate of Food and beverages and Pan, tobacco and intoxicants in Urban areas across different years for the month of January

````python
Urban_Jan_year=np.array(Urban_Jan['Year'])
fb=np.array(Urban_Jan['Food and beverages'])
pti=np.array(Urban_Jan['Pan, tobacco and intoxicants'])

fig,at = plt.subplots(figsize=(8,6))
at.set_title('Urban Sector for Jan Month')

plt.scatter(Urban_Jan_year,fb,c='orange',marker="o",label='Food and beverages')
plt.scatter(Urban_Jan_year,pti,c='cyan',marker="o",label='Pan, tobacco and intoxicants')

for index in range(len(Urban_Jan_year)):
  at.text(Urban_Jan_year[index],fb[index],(Urban_Jan_year[index],fb[index]))
  at.text(Urban_Jan_year[index],pti[index],(Urban_Jan_year[index],pti[index]))

plt.xlabel('Years')
plt.ylabel('Cost(in rupees)')
plt.legend(loc='lower right')
plt.show()
````

**Inferences drawn from the above Scatter plot-**

---


The rate of both food and beverages and pan, tobacco and intoxicants have increased over the years. However, we see a significant rise in the price of latter which shows that the consumption of these products have increased significantly over time and govt has imposed high taxes on these items.

# Line Plot

---



Making the Line Plot for the rate of Milk and products and fruits in rural areas across different years for the month of December

````python
rural_dec_year=np.array(Rural_Dec['Year'])
milkProducts=np.array(Rural_Dec['Milk and products'])
fruits=np.array(Rural_Dec['Fruits'])

fig,at = plt.subplots(figsize=(8,6))
at.set_title('Rural Sector for Dec Month')

plt.plot(rural_dec_year,milkProducts,c='green',marker="X", markersize=8,label='Rate of Milk and products')
plt.plot(rural_dec_year,fruits,c='blue',marker='.', markersize=10,label='Rate of Fruits')

for index in range(len(fruits)):
  at.text(rural_dec_year[index],milkProducts[index],(rural_dec_year[index],milkProducts[index]))
  at.text(rural_dec_year[index],fruits[index],(rural_dec_year[index],fruits[index]))

plt.xlabel('Years')
plt.ylabel('Cost (in rupees)')
plt.legend(loc='lower right')
plt.show()
````

**Inferences drawn from the above line plot-**

---


The rate of milk and products shown a steady rising pattern till the year 2017 when the rate started decreasing for an year and thereafter started increasing slightly from 2018.
The rate of fruits also obsevred a similar pattern till the year 2017 after which the rates became stagnant for an year and thereafter again started increasing gradually.

# Bar Plot

---



````python
Urban_rural_Jul_year=np.array(Rural_Urban_Jul['Year'])
fuel_light=np.array(Rural_Urban_Jul['Fuel and light'])
trans_comn=np.array(Rural_Urban_Jul['Transport and communication'])

fig,at = plt.subplots(figsize=(10,4))
at.set_title('Rural+Urban Sector for Jul Month')
plt.bar(Urban_rural_Jul_year,fuel_light,color="yellow",label='Fuel and light')
plt.xlabel('Years')
plt.ylabel('Cost(in rupees)')

plt.plot(Urban_rural_Jul_year,trans_comn,c='red',label='Transport and communication')
for index in range(len(Urban_rural_Jul_year)):
  at.text(Urban_rural_Jul_year[index],fuel_light[index],(Urban_rural_Jul_year[index],fuel_light[index]),c="blue")
plt.legend()
plt.show()
````

**Inferences drawn from the above bar plot-**

---
The rate of Transport and communication has increased significantly over the years due to increase in the prices of fuel and light.
